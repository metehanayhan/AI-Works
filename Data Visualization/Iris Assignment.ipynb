```python
# define function that accepts a row of data and returns a predicted species
def tahmin(petal_width):
    if petal_width <= 0.7:
        return 'setosa'
    elif 0.9 <= petal_width <= 1.79:
        return 'versicolor'
    elif petal_width >= 1.8:
        return 'virginica'

# 5 deger iÃ§in yanlÄ±ÅŸ tahmin Ã¼retecek..
df['predicted_species'] = df['petal_width'].apply(tahmin)

df[['petal_width', 'predicted_species']]
```
