## Feature Engineering

```python
df.corr(numeric_only=True)
```

```python
abs(df.corr(numeric_only=True)["Price"].sort_values(ascending=False))
```

```python
#kolerasyonu 0.2 ve 0.9 arasında olanları model için kullanacağız diğerlerinin katkısı çok az çünkü..(0.9 üstü olanlar aynı şeyi temsil ettiği için farklı formlarda almamıza gerek yok..)
```

```python
x=df[['Mileage','Cylinder','Liter','Cruise']]
y=df[['Price']]  
```

```python
#yukardaki satırda features ve target olarak veriyi ikiye ayırdık.. y değeri hedefimiz oldu.. tahmin etmek istenen değer.. Şimdi dikey bölmüş olduk bir de yatay böleceğiz..
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42) #test boyutunu %20 ayarladık 80 oto. eğitim için atandı..
```

```python
x_train.shape, x_test.shape
```

```python
#Artık data hazırlandı.. şimdi modeling
```

## Modelling

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
model=lr.fit(x_train,y_train) #Makine Öğrenmesi
```

```python
tahmin= model.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
mean_squared_error(y_test,tahmin)**.5
```

```python
r2_score(y_test,tahmin)
```

```python
#Label Encoding
```

```python
x=df[['Mileage','Cylinder','Liter','Cruise','Make','Trim']]
y=df[['Price']]  
```

```python
x.head()
```

```python
#stringleri sparse matrix e dönüştürerek modele dahil etmeliyiz
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42) 
```

```python
model=lr.fit(x_train,y_train)
```

```python
tahmin= model.predict(x_test)
```

```python
mean_squared_error(y_test,tahmin)**.5  #hata payı
```

```python
r2_score(y_test,tahmin)
```
