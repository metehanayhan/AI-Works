### Internetten Veri Alma

```python
url='https://en.wikipedia.org/wiki/Natural_language_processing'
```

```python
#pip install requests
```

```python
import requests
```

```python
dir(requests)
```

```python
resp=requests.get(url)
```

```python
resp #200: olumlu, 401:olumsuz
```

```python
#resp.text #sayfanın tamamını verir html halini.
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(resp.text,'lxml')
```

```python
#soup
```

```python
for p in soup.find_all('p'):
    print(p.text)
```

```python
#for a in soup.find_all('a'):
    #print(a)  #parafrafları getirir..
```

```python
for a in soup.find_all('a'):
    print(a.text)
```
