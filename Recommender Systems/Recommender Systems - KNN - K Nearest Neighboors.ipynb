```python
movie_dict = {}
with open ('ml-100k/u.item', encoding='latin1') as f:
    for line in f:
        fields = line.rstrip('\n').split('|')
        movie_id = int(fields[0])
        name = fields[1]
        genres = fields[5:24]
        genres =list(map(int, genres)) 
        movie_dict[movie_id] = (name, genres, movie_properties.loc[movie_id].rating.get('size'), movie_properties.loc[movie_id].rating.get('mean') )
```

```python
#movie_dict
```

```python
from scipy import spatial
def compute_distance(a, b):
    genre_distance = spatial.distance.cosine(a[1], b[1])
    popularity_distance = abs(a[2] - b[2])
    return genre_distance + popularity_distance
```

```python
def knn_recommendation(movie_id, k=5):
    distances = []  # Mesafeleri tutacağımız liste
    target_movie = movie_dict[movie_id]  
    
    # Diğer filmlerle olan uzaklıkları hesaplayalım
    for movie in movie_dict:
        if movie != movie_id: 
            distance = compute_distance(target_movie, movie_dict[movie])
            distances.append((movie, distance))
    
    distances.sort(key=lambda x: x[1])
    nearest_neighbors = distances[:k]
    
    recommendations = [(movies[movies['movie_id'] == movie[0]]['title'].values[0], movie[1]) for movie in nearest_neighbors]
    
    return recommendations
```

```python
# Test edelim (Örnek: Movie ID 1 "Toy Story (1995)" için en yakın 5 film)
print("Toy Story için önerilen filmler:")
knn_recommendation(1, k=5)
```
